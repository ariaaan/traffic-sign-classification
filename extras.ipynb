```python
import os
import cv2
import numpy as np

%matplotlib inline
from matplotlib import pyplot as plt
```

```python
image = cv2.imread('/home/arian/test.jpg')
image = image[...,::-1]
```

```python
x1, y1, x2, y2 = 150, 350, 300, 425
```

```python
label = 'Maxima 60'
```

```python
coords = (x1, y1), x2-x1, y2-y1
```

```python
colors = plt.cm.hsv(np.linspace(0, 1, 21)).tolist()
color = colors[0]
```

```python
plt.imshow(image)
ax = plt.gca()
ax.add_patch(plt.Rectangle(*coords, fill=False, edgecolor=color, linewidth=2))
ax.text(x1, x2, label, bbox={'facecolor': color, 'alpha': 0.5})
plt.show()
```

```python
x1, x2, y1, y2
```

```python
width = x2-x1
height = y2-y1

print(width, height)
```

```python
center_x = x1 + int(width/2)
center_y = y1 + int(height/2)

print(center_x, center_y)
```

```python
percent = 0.2
```

```python
size = max(width, height)

print(size)
size_2 = int(size/2 * (1 + percent))
print(size_2)
```

```python
img_height, img_width, _ = image.shape

print(img_width, img_height)
```

```python
x1 = center_x - size_2
x2 = center_x + size_2
y1 = center_y - size_2
y2 = center_y + size_2

if x1 < 0:
    x1 = 0
if x2 > img_width:
    x2 = img_width
if y1 < 0:
    y1 = 0
if y2 > img_height:
    y2 = img_height

print(x1, y1, x2, y2)
```

```python
coords = (x1, y1), x2-x1, y2-y1
```

```python
plt.imshow(image)
ax = plt.gca()
ax.add_patch(plt.Rectangle(*coords, fill=False, edgecolor=color, linewidth=2))
ax.text(x1, x2, label, bbox={'facecolor': color, 'alpha': 0.5})
plt.show()
```
